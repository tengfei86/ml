```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_row',100)
pd.set_option('display.max_columns',200)
pd.set_option('display.float_format','{:20,.2f}'.format)
import os
import gc
import time
import sys
```

```python
df = pd.read_csv('./data/indiegogo_orders.csv',dtype  = {'Shipping Name':'str','Shipping Phone Number':'str','Shipping Address':'str','Shipping Address 2':'str','Shipping City':'str','Shipping State/Province':'str','Shipping Zip/Postal Code':'str','Shipping Country':'str'})
```

```python
 def drawbar(col):
    data = df[col].value_counts().reset_index().rename(columns = {col:'count'})
    sns.barplot(x = 'count',y = 'index',data = data[:25])
```

```python
drawbar('Shipping Country')
```

```python
drawbar('Shipping City')

```

```python
drawbar('Appearance')
```
